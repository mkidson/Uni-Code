```python
import ROOT as r
hasITS = './Data/ITS/AnalysisResults_pass4_TracksIU_hasITS.root'
nohasITS = './Data/ITS/AnalysisResults_pass4_TracksIU_nohasITS.root'
```

```python
f3 = r.TFile(hasITS)
f4 = r.TFile(nohasITS)

# f.GetListOfKeys().Print()
```

```python
histRegistryName = 'hist-registry'
histName = 'eta'

f3.cd(histRegistryName)
hasITSHist = r.gDirectory.Get(histName)


f4.cd(histRegistryName)
nohasITSHist = r.gDirectory.Get(histName)

r.gDirectory.GetListOfKeys().Print()
```

```python
title = '#eta from ITS'
x_axis_label = '#eta'
x_axis_center = False
y_axis_label = 'Counts'
y_axis_center = False
log_x = False
log_y = False
grid_x = False
grid_y = False
aspectRatio = 1.7
line_colour = '#0000FF'
line_colour2 = '#FF0000'
fill_colour = '#c6d5eb'
fill_colour2 = '#FFC7C9'

```

```python
c = r.TCanvas()
c.SetGridx(grid_x)
c.SetGridy(grid_y)
c.SetLogx(log_x)
c.SetLogy(log_y)

hasITSHist.SetStats(0)
hasITSHist.SetLineWidth(2)
hasITSHist.SetLineColor(r.TColor.GetColor(line_colour))
hasITSHist.SetFillColorAlpha(r.TColor.GetColor(fill_colour),0.4)

nohasITSHist.SetStats(0)
nohasITSHist.SetLineWidth(2)
nohasITSHist.SetLineColor(r.TColor.GetColor(line_colour2))
nohasITSHist.SetFillColorAlpha(r.TColor.GetColor(fill_colour2),0.4)

hasITSHist.SetFillStyle(3000)
nohasITSHist.SetFillStyle(3000)

leg = r.TLegend(0.1245111,0.626322,0.4243807,0.8354877)
leg.AddEntry(hasITSHist, 'hasITS')
leg.AddEntry(nohasITSHist, 'nohasITS')
leg.SetFillColorAlpha(1,0)
leg.SetLineColorAlpha(1,0)
leg.SetTextSize(0.04)


nohasITSHist.GetXaxis().SetTitle(x_axis_label)
nohasITSHist.GetXaxis().CenterTitle(x_axis_center)
nohasITSHist.GetXaxis().SetTitleSize(0.05)
nohasITSHist.GetXaxis().SetTitleOffset(0.7)


nohasITSHist.GetYaxis().SetTitle(y_axis_label)
nohasITSHist.GetYaxis().CenterTitle(y_axis_center)
nohasITSHist.GetYaxis().SetTitleSize(0.05)
nohasITSHist.GetYaxis().SetTitleOffset(0.75)

nohasITSHist.SetTitle(title)

nohasITSHist.Draw('same')
hasITSHist.Draw('same')
leg.Draw('same')

c.Draw()

# c.SaveAs('./testagain.cpp')
```

```python
saveName = 'pass3_pass4'
c.SaveAs(f'{saveName}.pdf')
```
