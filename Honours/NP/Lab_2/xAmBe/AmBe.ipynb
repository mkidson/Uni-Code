```python
import numpy as np
from scipy.optimize import curve_fit
import scipy.signal
import scipy.stats
import matplotlib.pyplot as plt
import mplhep
import pandas as pd
```

```python
ambe_df = pd.read_csv('AmBe_good.csv', skiprows=21)
```

```python
channels = ambe_df['Channel'].to_numpy()
ambe_counts = ambe_df['Counts'].to_numpy()

ambe_counts = ambe_counts / 12906.99
```

```python
calibrationVals = [0.19255936, -2.35811083]
channels_calib = (channels - calibrationVals[1]) / calibrationVals[0]
```

```python
plt.figure(figsize=(10,5))
plt.hist(channels_calib, bins=channels_calib, weights=ambe_counts, histtype='step')
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.grid(color='#CCCCCC', linestyle=':')
```
