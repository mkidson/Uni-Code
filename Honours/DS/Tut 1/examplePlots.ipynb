```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
plt.figure()
x = np.linspace(-2,2,100)
y = x**2

z = np.polyfit(x, y, 2)
p = np.poly1d(z)

y0 = 1.0
crit = (p - y0).roots # roots of the polynominal -1, i.e., the mz values where p = 1 

print(crit)

#shading in the uncertainty band 
px=np.arange(crit[1],crit[0],0.001)
fig, ax = plt.subplots()

ax.fill_between(px,p(px),alpha=0.5, color='g', label="68% Confidence Interval")
ax.legend()
ax.set_xlim(-2.0, 2.0)

plt.plot(x, p(x), 'k', linewidth=2)

plt.hlines(1.0, -2.5, 2.0, colors='k', linestyles='--', label='')

plt.xlabel("theta")
plt.ylabel("delta chi-squared")
plt.show()
```
