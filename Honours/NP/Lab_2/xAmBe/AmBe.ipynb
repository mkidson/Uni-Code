```python
import numpy as np
from scipy.optimize import curve_fit
import scipy.signal
import scipy.stats
import matplotlib.pyplot as plt
import mplhep
import pandas as pd
# %matplotlib qt
```

```python
ambe_df = pd.read_csv('AmBe_good.csv', skiprows=21)
```

```python
channels = ambe_df['Channel'].to_numpy()
ambe_counts = ambe_df['Counts'].to_numpy()

ambe_counts = ambe_counts / 12906.99
```

```python
calibrationVals = [0.19255936, -2.35811083]
channels_calib = (channels - calibrationVals[1]) / calibrationVals[0]
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('Energy (keV)', fontsize=15)
plt.ylabel('Counts', fontsize=15)
plt.grid(color='#CCCCCC', linestyle=':')

plt.axvline(2230, color='C1', label='2224.497 keV (2H*)', lw=1.5)
plt.axvline(2559, color='C2', label='Unknown ~2560 keV', lw=1.5)
plt.axvline(3420, color='C3', label='Double escape peak', lw=1.5)
plt.axvline(3920, color='C4', label='Single escape peak', lw=1.5)
plt.axvline(4440, color='C5', label='4439.82 keV (12C*)', lw=1.5)


plt.hist(channels_calib, bins=channels_calib, weights=ambe_counts, histtype='step', label='Gamma Spectrum')

plt.legend(loc='upper right', fontsize=14)
plt.savefig('../Report/Plots/ambe.pdf')
```
